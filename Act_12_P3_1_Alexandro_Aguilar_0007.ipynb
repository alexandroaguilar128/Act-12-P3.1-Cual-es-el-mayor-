```python
# Para cuando es verdadero (V)
# Determina el numero mayor
# Datos de entrada A, B
# Mostrar B si es falso
# Si es verdadero mostrar A
A = int(input("Dame el primer numero: "))
B = int(input("Dame el segundo numero: "))
if A > B:
  print("Es verdadero")
  M = A > B
else:
  print("Es falso")
  M = A < B
  print("Fin del programa")
print(M)
print("Programa realizado por Alexandro Aguilar NL = 1")
```

```python
# Para cuando es Falso (F)
# Determina el numero mayor
# Datos de entrada A, B
# Mostrar B si es falso
# Si es verdadero mostrar A
A = int(input("Dame el primer numero: "))
B = int(input("Dame el segundo numero: "))
if A > B:
  print("Es verdadero")
  print(f"El numero mayor es: {A}")
  print("Fin del programa")
else:
  print("Es falso")
  print(f"El numero mayor es: {B}")
  print("Fin del programa")
print("Programa realizado por Alexandro Aguilar NL = 1")
```
